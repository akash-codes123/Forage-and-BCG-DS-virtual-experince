
To verify the hypothesis of price sensitivity being correlated with churn, we can define price sensitivity as the percentage change in consumption for a given percentage change in price. 

```python
import pandas as pd

# Calculating price sensitivity for each i.e peak , offpeak and midpeak.
df1['consumption_change_off_peak'] = (df1['cons_12m'] - df1['cons_last_month']) / df1['cons_last_month']
df1['price_change_off_peak'] = (df2['price_off_peak_var'] - df2['price_off_peak_fix']) / df2['price_off_peak_fix']
df1['price_sensitivity_off_peak'] = df1['consumption_change_off_peak'] / df1['price_change_off_peak']

df1['consumption_change_peak'] = (df1['cons_12m'] - df1['cons_last_month']) / df1['cons_last_month']
df1['price_change_peak'] = (df2['price_peak_var'] - df2['price_peak_fix']) / df2['price_peak_fix']
df1['price_sensitivity_peak'] = df1['consumption_change_peak'] / df1['price_change_peak']

df1['consumption_change_mid_peak'] = (df1['cons_12m'] - df1['cons_last_month']) / df1['cons_last_month']
df1['price_change_mid_peak'] = (df2['price_mid_peak_var'] - df2['price_mid_peak_fix']) / df2['price_mid_peak_fix']
df1['price_sensitivity_mid_peak'] = df1['consumption_change_mid_peak'] / df1['price_change_mid_peak']

# Calculating correlation between price sensitivity and churn
correlation = df1[['price_sensitivity_off_peak', 'price_sensitivity_peak', 'price_sensitivity_mid_peak', 'churn']].corr()
correlation_with_churn = correlation['churn']

print(correlation_with_churn)

```
